## Length of the sequence $l$

```python
l=10
```

```python
def verify(n):
    if sum(n) == 0:
        partial = 0
        for i in n:
            partial = partial + i
            if partial < 0:
                return False
        return True
    return False
  
        
def prob(x):
    count = 0
    for _ in range(x):
        seq = [random.choice([-1,1]) for _ in range(l)]   
        if verify(seq):
            count += 1
        
    return count/x





    

        
```

We define two functions **verify** (*verfying our two conditions*) and **prob** (*computing and counting the verified sequences*)
